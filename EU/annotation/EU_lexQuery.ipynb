```python
#save html and pdf content differently!
#Documents
# Define the folder paths where you want to save the files
from_html_path = r'C:\\Users\\cecilianatalie.strom\\Documents\\EU_lex\\from_html'
from_pdf_path = r'C:\\Users\\cecilianatalie.strom\\Documents\\EU_lex\\from_pdf'

# Create the folders if they don't exist
os.makedirs(from_html_path, exist_ok=True)
os.makedirs(from_pdf_path, exist_ok=True)


target_sentence = '\n The requested document does not exist.\n'

# Function to fetch HTML content
def fetch_html_content(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for HTTP errors
        return response.text
    except requests.RequestException as e:
        print(f"An error occurred while fetching HTML content from {url}: {e}")
        return None

# Function to fetch and save PDF content
def fetch_pdf_content(pdf_url, output_file):
    try:
        response = requests.get(pdf_url)
        response.raise_for_status()  # Raise an exception for HTTP errors
        with open(output_file, 'wb') as f:
            f.write(response.content)
        print(f"PDF content saved as '{output_file}'.")
    except requests.RequestException as e:
        print(f"An error occurred while fetching PDF content from {pdf_url}: {e}")

# Check each URL and save the content appropriately
for i, site in enumerate(websites):
    html_content = fetch_html_content(site)
    if html_content is not None:
        if target_sentence in html_content:
            print(f"The requested document does not exist at: {site}, fetching PDF instead.")
            pdf_url = site.replace('HTML', 'PDF')
            pdf_output_file = os.path.join(from_pdf_path, f'document_{i + 1}.pdf')
            fetch_pdf_content(pdf_url, pdf_output_file)
        else:
            print(f"Content available at: {site}")
            html_output_file = os.path.join(from_html_path, f'document_{i + 1}.html')
            with open(html_output_file, 'w', encoding='utf-8') as file:
                file.write(html_content)
            print(f"Saved HTML content from {site} to {html_output_file}")
```
